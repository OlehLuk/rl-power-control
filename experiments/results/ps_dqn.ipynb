```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def no_control_performance(path, p_ref):   
    df = pd.read_csv(path, header=None)
    plt.plot(df[5])
    diffs = (df - p_ref).values
    mses = np.mean(np.power(diffs, 2), axis=0)
    print("Mean MSE {}".format(np.mean(mses)))
    print("Std MSE {}".format(np.std(mses)))
    print("Medin MSE {}".format(np.median(mses)))
    return mses
```

```python
no_control_performance("ps_stoch/new_compete/uss_no_control.csv", 1.2)
```
