# Action space variation experiment for deterministic model, constant p_reff

```python
df_ks_1_avgsm_mses, perf_ks1, sig_ks1 = report_experiment(
    "./ps_det/action_space(k_s)_variation_19-10-2019_22-08/k_s=[1, 2, 3]", 
    exp_name="\nDet.PS, Q-learning, Action space variation: k_s=[1,2,3]", stochastic=False)
```

```python
df_ks_2_avgsm_mses, perf_ks2, sig_ks2 = report_experiment(
    "./ps_det/action_space(k_s)_variation_19-10-2019_22-08/k_s=[0.5, 1, 2, 3, 4]", 
    exp_name="\nDet.PS, Q-learning, Action space variation: k_s=[0.5, 1, 2, 3, 4]", stochastic=False)
```

```python
df_ks_3_avgsm_mses, perf_ks3, sig_ks3 = report_experiment(
    "./ps_det/action_space(k_s)_variation_19-10-2019_22-08/k_s=[0.1, 0.5, 1, 2, 7]", 
    exp_name="\nDet.PS, Q-learning, Action space variation: k_s=[0.1, 0.5, 1, 2, 7]", stochastic=False)
```

```python
labels = ["[1,2,3]", "[0.5,1,2,3,4]", "[0.1,0.5,1,2,7]"]
mses = [df_ks_1_avgsm_mses, df_ks_2_avgsm_mses, df_ks_3_avgsm_mses]
perfs = [perf_ks1, perf_ks2, perf_ks3]
sigs = [sig_ks1, sig_ks2, sig_ks3]
report_experiment_summary("Different action spaces", mses, perfs, sigs, labels, 0)
```
